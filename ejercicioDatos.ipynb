```python
import matplotlib.pyplot as plt

notas_matematica = ['Matemtica',8,7,6,6,7,7,8,10]
notas_lenguaje = ['Lenguaje',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografa',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_lenguaje, notas_geografia]

for nota in notas:
 x = list(range(1, 9))
 y = nota[1:]
 plt.plot(x, y, marker='o')
 plt.xlabel('Exmenes')
 plt.ylabel('Notas')
 plt.show()
```
